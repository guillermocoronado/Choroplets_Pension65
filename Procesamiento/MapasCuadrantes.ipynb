```python
import pandas as pd
DatosProvincias=pd.read_excel(os.path.join("Datos_por_zona", "DatosProvinciales", "DatosProvinciasv2.xlsx"))
```

```python
PeruProvincias.rename(columns = {'NOMBDEP':'Departamento'}, inplace = True)
PeruProvincias.head()
```

```python
PeruProvincias.rename(columns = {'NOMBPROV':'Provincia'}, inplace = True)
PeruProvincias.head()
```

```python
PeruProvincias.info()
```

```python
DatosProvincias.info()
```

```python
PeruMapaCuadrantes = PeruProvincias.merge(DatosProvincias,
                                           how='inner',left_on='Provincia',right_on='Provincia')
```

```python
PeruMapaCuadrantes
```

```python
PeruMapaCuadrantes.info()
```

```python
import matplotlib.pyplot as plt

f, ax = plt.subplots(1, figsize=(10, 10))
PeruMapaCuadrantes.plot(column='cuadrante2022', 
        cmap='viridis', 
        categorical=True,
        edgecolor='white', 
        linewidth=0., 
        alpha=0.75, 
        legend=True,
        legend_kwds={'loc':3},
        ax=ax
       )

ax.set_axis_off()
```

```python
PeruMapaCuadrantes.to_file(os.path.join("MapaCuadrantes.gpkg"), layer='Provincias', driver="GPKG")
```

```python
# facetting
f, axs = plt.subplots(nrows=2, ncols=3, figsize=(8, 8))
axs = axs.flatten()

for i in range(5):  
    ax = axs[i]# select the axis where the map will go
    PeruMapaCuadrantes.plot(ax=ax,color='gainsboro')
    PeruMapaCuadrantes[PeruMapaCuadrantes.cuadrante2022==i].plot(ax=ax,color='green')
    ax.set_axis_off()
    ax.set_title("level_"+str(i))
f.delaxes(axs[5]) # bye empty subplot
plt.show()
```
